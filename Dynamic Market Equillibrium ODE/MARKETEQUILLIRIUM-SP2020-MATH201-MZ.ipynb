 ### <font color=#0052cc ><b>First:Impoting Needed Libraries and Initializing the Equations and Variables:</b></font>

```python
#importing needed libraries
import matplotlib.pyplot as plt
import matplotlib as matplt
from matplotlib import style
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline

import numpy as np
import sympy as sym
import cmath
import math as m

#from gekko import GEKKO
#m = GEKKO()
#m.time = np.linspace(0,10)
```

```python
sym.init_printing()
t, w1 , u1 , v1 , c1 , w2 , u2, v2 , c2 ,W , U , V , C = sym.symbols('t w1 u1 v1 c1 w2 u2 v2 c2 w u v c')
S= sym.Function('S')(t)
D= sym.Function('D')(t)
P= sym.Function('P')(t)
dpdt = P.diff(t)
d2dpdt2= dpdt.diff(t)

expr1 = sym.Eq(S, c1 + w1*P + u1*dpdt + v1*d2dpdt2)
expr2 = sym.Eq(D, c2 + w2*P + u2*dpdt + v2*d2dpdt2)
expr3 = sym.Eq(((P)* (W/V))+((dpdt)*(U/V))+(d2dpdt2)  ,-(C/V))
expr1,expr2,expr3
```

<font family="Arial" color="Brown" ><b><center>Where W = Abs(W1 -W2), U = Abs(U1-U2), V = Abs(V1-V2) and C = Abs(C1-C2)</center></b></font>

## <font size=11 color = #660033><b> The Solving Phase:</b></font>

 ### <font color=#0052cc ><b>First: Getting Equilibrium Coefficients Values:</b></font>

```python
Wv=abs(float(w1v)-float(w2v))
print('W = ',Wv)
W=sym.symbols(str(Wv))

Uv=abs(float(u1v)-float(u2v))
print('U = ',Uv)
U=sym.symbols(str(Uv))

Vv=abs(float(v1v)-float(v2v))
print('V = ',Vv)
V=sym.symbols(str(Vv))

Cv=abs(float(c1v)-float(c2v))
print('C = ',Cv)
C=sym.symbols(str(Cv))
```

```python
expr3 = sym.Eq((P*(W))+((dpdt)*(U))+(d2dpdt2) ,-(C))
expr3
```

```python
expr3 = sym.Eq(((P)*(Wv/Vv))+((dpdt)*(Uv/Vv))+(d2dpdt2),-(Cv/Vv))
expr3
```

```python
Pp=sym.symbols('P0')
expr4=sym.Eq(Pp,-(C/W))

expr4
```

```python
P0=-(Cv/ Wv)
print('Po=',P0)

a=-(Uv/(2*Vv))
#print('a=',a)
b=(0.5)*(m.sqrt(4*(Wv/Vv)-((Uv/Vv)**2)))
#print('b=',b)
```

 ### <font color=#0052cc ><b>Second: Checking The Solution Case:</b></font>

```python
x=(-((Uv/Vv)/2))
R =(((Uv / Vv)**2)-(4* Wv/Vv))
#R=8
#R=0
A1 , A2 =sym.symbols('A1  A2')
print("Sqrt(((b)^2)-4ac)=",R)
expr5 = 0
if R > 0 :  #Real Roots
    print("Using Real Roots Case:")
    r1=round(x+(m.sqrt(R)/2),8)
    print("r1=",r1)
    r2=round(x-(m.sqrt(R)/2),8)
    print("r2=",r2)
    e1=sym.exp(r1*t)
    e2=sym.exp(r2*t)
    expr5 = sym.Eq(P , A1*e1 + A2*e2 -P0)
elif R == 0  :
    print("Using Repeated root Special Case:") # r=x
    e=sym.exp(x*t)
    expr5 = sym.Eq(P , A1*e + A2*e - P0)
    

elif R<0 : #Complex Roots
    print("Using Complex Roots Case:")
    sin = sym.sin(b*t)
    cos = sym.cos(b*t)
    e=sym.exp(a*t)
    expr5 = sym.Eq(P , (e*(A1*cos + A2*sin )) - P0)
    
expr5  
```

```python
#QAns="y"
#while QAns =="y":
   # MNo1v=input("first month in no.: ")
   # Mp1v=input("price at first month: ")
   # MNo2v=input("2nd month in no.: ")
   # Mp2v=input("price at 2nd month: ")
   # QAns=input("Would You like to resubmit values?(y/n)")
MNo1v=str(1) 
Mp1v=str(186000)
MNo2v=str(2)
Mp2v=(193000)
exp1=expr5.subs(t,MNo1v)
exp1
```

```python
exp2=expr5.subs(t,MNo2v)
exp2
```

```python
eq1= expr5.subs( {P :(Mp1v),t :(MNo1v) })
eq1
```

```python
eq2= expr5.subs( {P :(Mp2v),t :(MNo2v) })
eq2
```

```python
sol= sym.solve((eq1,eq2),(A1,A2))
sol
```

```python
solA1=sol[A1]
solA2=sol[A2]
exprF=expr5.subs({A1 : solA1 , A2 : solA2 })
exprF
```

```python
time  = np.linspace(0,36,160, dtype = float)
x_vals = time
```

```python
X=exprF.rhs
X

```

 ### <font color=#0052cc ><b>Finally:Graphing Functions (SHOW TIME🎉):</b></font>

```python
lam_x = sym.lambdify(t, X, modules=['numpy'])
y_vals = lam_x(x_vals)
plt.figure(figsize = (13, 13))
plt.title("Graph P(t),price, over time t")
plt.plot(x_vals, y_vals)
plt.ylabel("price")
plt.xlabel("time")
plt.show()
```
